```python
file_path = "/mntssd/mnt3/shanshanbai/my_storage_from_qian/results/best_buildings.jsonl"
df = pd.read_json(file_path, lines=True)
len(df)
```

```python
df_commercial= df[df['building class'] == 'commercial']
df_residential= df[df['building class'] == 'residential']
```

```python
# Explode 'building names' to handle lists of names
df_exploded = df_residential.explode('building names')
```

```python
# Define a regex pattern for valid building names (words)
valid_name_pattern = r'^[a-zA-ZÀ-ÖØ-öø-ÿ\s\'-]+$'  # Allows letters, spaces, hyphens, and apostrophes

# Filter rows where 'building names' match the valid pattern
df_exploded = df_exploded[df_exploded['building names'].str.match(valid_name_pattern, na=False)]
```

```python
df_exploded["building tags"] = df_exploded["building tags"].apply(
    lambda tags: [tag for tag in tags if tag not in ["residential", "commercial"]]
)
df_exploded
```

```python
df_exploded = df_exploded.explode('building tags')
df_exploded
```

```python
# Language mapping dictionary
language_mapping = {
    'ar': 'Arabic', 'bg': 'Bulgarian', 'bn': 'Bengali', 'ca': 'Catalan', 'cs': 'Czech',
    'cy': 'Welsh', 'da': 'Danish', 'de': 'German', 'el': 'Greek', 'en': 'English',
    'es': 'Spanish', 'et': 'Estonian', 'eu': 'Basque', 'fa': 'Persian', 'fi': 'Finnish',
    'fr': 'French', 'hi': 'Hindi', 'ht': 'Haitian Creole', 'hu': 'Hungarian', 'hy': 'Armenian',
    'in': 'Indonesian', 'is': 'Icelandic', 'it': 'Italian', 'iw': 'Hebrew', 'ja': 'Japanese',
    'ka': 'Georgian', 'ko': 'Korean', 'lt': 'Lithuanian', 'lv': 'Latvian', 'mr': 'Marathi',
    'nl': 'Dutch', 'no': 'Norwegian', 'pa': 'Punjabi', 'pl': 'Polish', 'pt': 'Portuguese',
    'ro': 'Romanian', 'ru': 'Russian', 'sl': 'Slovenian', 'sr': 'Serbian', 'sv': 'Swedish',
    'th': 'Thai', 'tl': 'Tagalog', 'tr': 'Turkish', 'uk': 'Ukrainian', 'ur': 'Urdu',
    'vi': 'Vietnamese', 'zh': 'Chinese'
}

df_exploded["tweet language distribution"] = df_exploded["tweet language distribution"].apply(
    lambda langs: [language_mapping.get(lang, lang) for lang in langs]
)
df_exploded
```

```python
# Truncate lists in the 'tweet language distribution' column
df_exploded["tweet language distribution"] = df_exploded["tweet language distribution"].apply(lambda x: x[:5])
df_exploded
```
